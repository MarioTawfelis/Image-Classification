### Helper functions

```python
# This function creates directories for each person in our dataset
# These directories will then be used to store cropped images
# Source: https://github.com/codebasics/py/blob/master/DataScience/CelebrityFaceRecognition/model/data_cleaning.ipynb
def get_image_dirs():
    img_dirs = []
    for entry in os.scandir(path_to_data):
        if entry.is_dir():
            img_dirs.append(entry.path)
    return img_dirs
```

```python
# We need to automate this process to batch process our dataset
# Validity of an image is if the image contains visible, unobstructed eyes
def get_cropped_image_if_valid(image_path):
    img = cv.imread(image_path)
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
cropped_img = get_cropped_image_if_valid('./datasets/beyonce/0f80c7cb9f.jpg')
plt.imshow(cropped_img)
```

## Data Cleaning

```python
# Create required variables
cropped_image_dirs = []
leader_file_names = {}
```

```python
# Retrieve names of directories
img_dirs = get_image_dirs()
img_dirs
```

```python
# 1 - Iterate over each image directory in our dataset
for img_dir in img_dirs:
    counter = 1
    
    # 2 - Retrieve person name from the directory name
    leader_name = img_dir.split('/')[-1]
    print(leader_name)
    
    leader_file_names[leader_name] = []
    
    # 3 - Iterate over each image in the current directory
    for entry in os.scandir(img_dir):
        
        # 4 - Detect faces and return cropped image if valid (i.e. face contains 2 visible eyes)
        roi_color = get_cropped_image_if_valid(entry.path)
        
        # None is returned in case of an invalid face
        if roi_color is not None:
            # 5 - Create new folder to store valid cropped images
            cropped_folder = path_to_cr_data + leader_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
                
            # 6 -  Create file names for cropped images
            cropped_file_name = leader_name + str(counter) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name 
            
            # 7 - Store cropped image
            cv.imwrite(cropped_file_path, roi_color)
            leader_file_names[leader_name].append(cropped_file_path)
            counter += 1

            
```

```python
# Clean our leader_file_names dictionary to remove references of manually deleted images
leader_file_names_dict = {}
for img_dir in cropped_image_dirs:
    leader_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(img_dir):
        file_list.append(entry.path)
    leader_file_names_dict[leader_name] = file_list
leader_file_names_dict
```

### Feature Engineering

For our classification model, our target variable is the name of a female leader of the 21st century. Hence, we need to encode each name to a number.

```python
# Importing required packages
import pywt
```

#### Wavelet Transfor

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor(imArray,cv2.COLOR_RGB2GRAY)
    
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    
    # compute coefficients 
    coeffs = pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H = pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
# Encoding names to numbers
class_dict = {}
counter = 0
for leader_name in leader_file_names_dict.keys():
    class_dict[leader_name] = counter
    counter = counter + 1
class_dict
```

```python
# Define our independent and target variables
X, y = [], []
```

```python
for leader_name, training_files in leader_file_names_dict.items():
    for training_image in training_files:
        # 1 - Read raw image
        img = cv.imread(training_image)
        
        if img is None:
            continue
            
        # 2 - Rescale raw image to 32x32 
        scalled_raw_img = cv.resize(img, (32, 32))
        
        # 3 - Create a WT image
        img_har = w2d(img,'db1',5)
        
        # 4 - Rescale the transformed image to 32x32
        scalled_img_har = cv.resize(img_har, (32, 32))
        
        # 5 - Vertically stack the raw and transformed images 
        stacked_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        
        # 6 - Contruct independent and target variables
        X.append(stacked_img)
        y.append(class_dict[leader_name])     
```

```python
X = np.array(X).reshape(len(X),4096).astype(float)
X.shape
```

### Model Training

For this porject, I am using SVM for classification and GridSearch to tune the hyperparameters.

```python
# Import required packages
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# Split our data into train/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
# Create a simple pipeline to normalize and train our model
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC(kernel = 'rbf', C = 10))])
```

```python
# Train SVM model
pipe.fit(X_train, y_train)
```

```python
# Get an initial accuracy score
pipe.score(X_test, y_test)
```

```python
# Get a detailed classification report
print(classification_report(y_test, pipe.predict(X_test)))
```
